```python
#Importar las libreria pandas, numpy, seaborn, matplotlib 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#Leer la base de datos genres_v2.csv a través de la libreria pandas.
#Asignaremos el nombre spotify a la base de datos. 
#Desplegamos datos iniciales y finales de la base de datos con el comando "spotify"
spotify= pd.read_csv("genres_v2.csv")
spotify
```

```python
#El comando spotify.shape nos muestra la cantidad de columnas y filas en la base de datos 
spotify.shape
```

```python
#spotify.head() nos muestra las primeras 5 filas de datos 
spotify.head()
```

```python
#spotify['duration_ms'] nos centra en una columna especifica de la duración de las canciones.
#Nombramos a esta columna como sp
#Realizamos un analisis estadistico de la duración de canciones, con media, mediana, maximo y minimo.
sp=spotify['duration_ms']
sp.mean(), sp.median(), sp.min(), sp.max()
```

```python
#Filtrando por genero musical, spotify['genre'] == 'hardstyle' nos muestra cuantos generos son efectivamente harstyle. 
spotify['genre'] == 'hardstyle'
```

```python
#Descripción del genero "Hardstyle", esta función nos muestra la descripción general de todos las canciones con genero hardstyle.
spotify[spotify['genre'] == 'hardstyle']
```

```python
#Resumen general estadistico de la base de datos.
spotify.describe()
```

```python
#Probailidad de escoger una canción de x duración. 
sns.histplot (x=sp, bins=30, kde=True, stat='probability', color='blue' )
```

```python
#HIstograma de la columna de duracion--> Aqui nos muestra cual es la duracion mas usada para en general las canciones que se encuentran
# en spotify
histogram_duration_ms= sns.histplot(x=spotify["duration_ms"])
```

```python
#boxplot de la duracion de las canciones, no en general, si no especificamente por genero musical
boxplot_duration_ms= sns.boxplot(x=spotify["duration_ms"], y=spotify["genre"], showmeans=True)
```

```python
#Boxplot general de la duracion con datos maximos, minimos y media de las canciones registradas sin identificar
#genero musical u otra variable
boxplot_duration_ms= sns.boxplot(x=spotify["duration_ms"], showmeans=True)
```

```python
#Diagrama de densidad--> Diagrama de densidad de cada genero musical con respecto a la duración, de los diferentes 
#generos como trap, techno, RnB, pop, hip hop... entre otros, todo con respecto a la cantidad de canciones que respectivamente
#hay en cada genero
sns.displot(x="duration_ms",row="genre",data=spotify, linewidth=1,kde=True)
```

```python
#scatterolot--> Donde se relacionan mas la cantidad de canciones con respecto a la duracion y la bailabilidad, en las zonas 
#en las zonas donde hay mas puntos significa que hay mas canciones que duran una media de 2000000 ms y con una bailabilidad de 0.6
scatter_plot= sns.scatterplot(x="duration_ms", y="danceability",data=spotify)
scatter_plot.set_title("Duration vs danceability")
scatter_plot.set_xlabel("duration_ms")
scatter_plot.set_ylabel("danceability")
```

```python
#este diagrama de barras relaciona el genero con la bailabilidad 
plt.bar("genre","danceability", data = spotify, color = "green")
plt.xlabel("genre")
plt.ylabel("danceability")
plt.title("Genre vs danceability")
plt.show()
```

```python
#Diagrama de pie-->AL ser tantos datos el diagrama no puede concretarse en un patron solido entre la energia y bailabilidad
plt.pie("energy","danceability", data = spotify,)
plt.show()
```

```python
#kdeplot es como un relieve que relaciona la duración y la bailabilidad, en la zona en donde la duracion es de 200000 ms 
# y la danzabilidad de 0.6-0.5 es mayor
sns.kdeplot(x="duration_ms", y="danceability", data=spotify);
```

```python
# Scatterplot con variable numerica y de caracteres--> En donde se puede evidenciar  que nos dicen la duracion de las canciones por genero
scatter_plot= sns.scatterplot(x="duration_ms", y="genre",data=spotify)
scatter_plot.set_title("Duration vs genre")
scatter_plot.set_xlabel("duration_ms")
scatter_plot.set_ylabel("genre")
```

```python
#Overlapped chard--> Cantidad de energia empleada por genero musical

sns.histplot(x = 'energy', hue = 'genre', data = spotify, linewidth=1)
```
